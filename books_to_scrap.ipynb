```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
r = requests.get('https://books.toscrape.com/')
r
```

```python
ratings = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}
```

```python
# get name, price and rating

soup = BeautifulSoup(r.text, 'html.parser')

books = soup.find_all("article", attrs={"class":"product_pod"})
name = books[0].find_all("a")[1].get('title')
price = float(books[0].find('p', attrs={"class":"price_color"}).get_text()[2:])
rating = books[0].find('p', attrs={"class":'star-rating'}).get('class')[1]
```

```python
# get category

all_cats = soup.find_all('ul', class_= 'nav nav-list')[0]
li = all_cats.find_all('li')[1:]
li
```

```python
# get category name
category = li[1].get_text().split()
cat_name = ' '.join(category)
cat_name
```

```python
with open("books.csv", 'w', encoding = 'utf-8', newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Category", "Name", "Price", "Rating"])   
    
    for link in li:
        
        hlink = link.find('a').get('href')
        new_link = 'https://books.toscrape.com/'+ hlink
        category = link.get_text().split()
        cat_name = ' '.join(category)
        
        response = requests.get(new_link)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        books = soup.find_all("article", attrs={"class":"product_pod"})
        for article in books:
            
            name = article.find_all("a")[1].get('title')
            price = float(article.find('p', attrs={"class":"price_color"}).get_text()[2:])
            rating = article.find('p', attrs={"class":'star-rating'}).get('class')[1]
            writer.writerow([cat_name, name, price, rating])
        
        while True:
            try:
                
                pages = soup.find('li', class_='next').find('a',href = True)['href']
                print(pages)
                next_page = new_link.split('/')
                next_page[-1] = pages
                next_page = '/'.join(next_page)
                next_page
                r = requests.get(next_page)
                soup = BeautifulSoup(r.text,'html.parser')
                books = soup.find_all("article", attrs={"class":"product_pod"})
                
                for nxt in books:
                    
                    name = nxt.find_all("a")[1].get('title')
                    price = float(nxt.find('p', attrs={"class":"price_color"}).get_text()[2:])
                    rating = nxt.find('p', attrs={"class":'star-rating'}).get('class')[1]
                    
                    writer.writerow([cat_name, name, price, rating])
                
            except Exception as e:
                break
            
```
